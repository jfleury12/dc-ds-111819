Write a function to find all possible unique permutations of the word 'Mississippi'.
word = 'Mississippi'
Bonus: Write the function in such a way that it can find all possible unique permutations of any word.


```python
import itertools
```

```python
word = 'Mississippi'

print(list(itertools.permutations([7, 3, 0, 0])))
set(itertools.permutations([7, 3, 0, 0]))
```

```python
x = list(itertools.permutations('Mississippi'))
```

```python
print(len(x))
set(x)
```

```python
def unique_perm(word):
    return len(set(itertools.permutations(word)))

unique_perm('Mississippi')
```
