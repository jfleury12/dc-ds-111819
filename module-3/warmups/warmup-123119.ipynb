```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
np.random.seed(2019)
```

```python
x_norm = np.random.normal(10,1.5,1000)
```

```python
# obtain the means of 1000 random samples from a normal distribution
means = np.zeros(1000)

for n in range(1000):
    means[n] = np.random.normal(10,1.5,1000).mean()
```

```python
sns.distplot(x_norm)
sns.distplot(means)
plt.title('Normal Distribution Overlayed')
# narrow, tighter normal distribution around mean
```

**Bonus: Repeat the above with the Chi-Squared distribution**

```python
chi2 = np.random.chisquare(4,1000)
```

```python
chi_means = np.zeros(1000)

for n in range(1000):
    chi_means[n] = np.random.chisquare(4,1000).mean()
```

```python
sns.distplot(chi2)
sns.distplot(chi_means)
plt.title('Chi-Squared Distribution Overlay')
# narrow, tighter normal distribution around mean
```
