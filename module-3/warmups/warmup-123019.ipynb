Set random seed to 2019.
- Randomly sample 1000 instances from the Normal Distribution with mean 10 and standard deviation 1.5 and store them in a variable called x_norm. 
- Plot a histogram of x_norm.
- Then, create a variable called means and store the mean obtained from 1000 random samples independent normal Normal Distributions, each with mean 10 and standard deviation 1.5. 
- Plot a histogram of means. 
- Post your two histogram plots (one of x_norm and one of means) in the thread.

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
np.random.seed(2019)
```

```python
x_norm = np.random.normal(10,1.5,1000)
```

```python
sns.distplot(x_norm)
plt.title('Sampled Normal Distribution')
```

```python
# obtain the means of 1000 random samples from a normal distribution

means = np.zeros(1000)
for n in range(1000):
    means[n] = np.random.normal(10,1.5,1000).mean()
```

```python
sns.distplot(means)
plt.title('Sample Means')
```
